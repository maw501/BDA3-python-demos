### Bayesian data analysis
##  Chapter 9, demo 1

Prof Gelman has a jar of coins. He promises that if the students guess how many coins there are, they will get all the coins in
the jar. Students discuss and guess different values. Based on these they eventually present their uncertainty about the number of coins as a normal distribution $\mathcal{N}(160, 40)$. 

What value they should guess?

```python
m = 160
s = 40

x = np.arange(m - 3 * s, m + 3 * s + 1)
px = stats.norm.pdf(x, loc=m, scale=s)
xpx = x * px
```

If students just want to guess right, and they do not care how much money they'll get they should guess the most probable value.

```python
plt.plot(x, px)
plot_tools.modify_axes.only_x(plt.gca())
plt.xlim([x[0], x[-1]])
plt.ylabel("probability")
plt.xlabel("number of coins")

(h2,) = plt.plot([m, m], [0, stats.norm.pdf(m, loc=m, scale=s)])
plt.legend(
    (h2,),
    ("most probable value = {}".format(m),),
    loc="lower center",
    bbox_to_anchor=(0.5, 1.02),
);
```

Alternatively students might want to maximize the exepected utility of the number coins. Assume that utility of the money is linear.

```python
plt.style.use(plot_tools.custom_styles["gray_background"])

plt.plot(x, x)
plt.ylabel("utility if guess is correct")
plt.xlabel("guess")
plt.xlim([x[0], x[-1]])
plt.ylim([x[0], x[-1]]);
```

If students guess value $a$, given their estimate of the uncertainity, probability that they get a coins is $p(a)$, and expected utility is $a*p(a)$.

```python
plt.plot(x, xpx)
plt.ylabel("expected utility")
plt.xlabel("guess")
plt.ylim([0, plt.ylim()[1]])
plt.xlim([x[0], x[-1]])

# Compute the maximum of the expected utility
mi = np.argmax(xpx)
meu = xpx[mi]
meux = x[mi]
(h3,) = plt.plot([meux, meux], [0, meu])
plt.legend(
    (h3,),
    ("guess which maximises the expected utility = {}".format(meux),),
    loc="lower center",
    bbox_to_anchor=(0.5, 1.02),
);
```
